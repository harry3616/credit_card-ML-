```python
# use the normal equation to get the label_month5_payment
def normal_equation(data_x, data_y):
    #- Calculate the number of rows of data_x, labeled as m
    m = len(data_x)
    #- Add a spacer column to x
    #x = np.c_[np.ones((m,1)), data_x]
    
    #- Theta = (xTx)^(-1)xTy 
    xTx = data_x.transpose().dot(data_x)
    theta = np.linalg.inv(xTx).dot(data_x.transpose()).dot(data_y)
    return theta



x2 = np.array(X2)
y2 = np.array(data2['previous_payment_month4'])[:, np.newaxis]
theta_ne2 = normal_equation(x2, y2)
l5=theta_ne2.T.dot(x2.T)


label_month5_payment=l5.tolist()
#label_month5_payment 即為所求
print(label_month5_payment)
```
