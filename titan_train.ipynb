```python
#- Normal Equation
def normal_equation(data_x, data_y):
    #- Calculate the number of rows of data_x, labeled as m
    m = len(data_x)
    #- Add a spacer column to x
    #x = np.c_[np.ones((m,1)), data_x]
    
    #- Theta = (xTx)^(-1)xTy 
    xTx = data_x.transpose().dot(data_x)
    theta = np.linalg.inv(xTx).dot(data_x.transpose()).dot(data_y)
    return theta


#- preprocess data
x = np.array(X)
y = np.array(data['previous_payment_month4'])[:, np.newaxis]
```

```python
#Find theta using Normal Equation because parameter less than 1000
# each wight as parameter as follow 
theta_ne = normal_equation(x, y)
print('theta_ne ', theta_ne)
max(theta_ne)

```

```python
#  Find theta using Scikit-learn
# each wight as parameter as follow 
reg = LinearRegression().fit(x, y)
theta_scikit = np.c_[reg.intercept_, reg.coef_]
print('theta_scikit ', theta_scikit.transpose())
max(theta_ne)
```
